```python
import numpy as np
filename = "height-weight.csv"
raw_data = open(filename, 'rt')
data = np. loadtxt(raw_data, delimiter = ',')
```

```python
data.shape
```

```python
data[:10]
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
X = data[:, 0]
y = data[:, 1]
```

```python
print(X.shape)
print(y.shape)
```

```python
X_20 = X[:20]
y_20 = y[:20]

plt.scatter(X_20,y_20, color= 'blue', s=20)

plt.xlabel('Height')
plt.ylabel('Weight')

plt.show()
```

```python
X_train = X[:4500]
X_test =  X[4500:]
y_train = y[:4500]
y_test =  y[4500:]
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
X_train = X_train.reshape(-1, 1)
X_test =  X_test.reshape(-1, 1)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

```python
print('Coefficients:', model.coef_)
```

```python
plt.scatter(X_test, y_test, color ='blue')
plt.plot(X_test, y_predict, color = 'red', linewidth =3)

plt.xlabel('Height')
plt.ylabel('Weight')

plt.xticks()
plt.yticks()

plt.show()
```
